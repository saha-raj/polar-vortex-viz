```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import os
import numpy as np
from scipy.signal import savgol_filter

def find_jet_stream(wind_magnitude, lats, lons):
    # Focus on Northern Hemisphere band
    lat_mask = (lats >= 30) & (lats <= 70)
    nh_winds = wind_magnitude[lat_mask]
    nh_lats = lats[lat_mask]
    
    # Find maximum wind speed latitude for each longitude
    jet_lats = np.zeros(len(lons))
    for i in range(len(lons)):
        wind_profile = nh_winds[:, i]
        # Use weighted average around maximum
        weights = wind_profile**2  # Square to emphasize stronger winds
        jet_lats[i] = np.average(nh_lats, weights=weights)
    
    # Ensure continuity across prime meridian
    jet_lats = savgol_filter(jet_lats, window_length=51, polyorder=3)
    
    # Force matching ends for periodic boundary
    avg_end = (jet_lats[0] + jet_lats[-1])/2
    jet_lats[0] = avg_end
    jet_lats[-1] = avg_end
    
    return jet_lats

def find_jet_stream_with_uncertainty(wind_magnitude, lats, lons):
    # Focus on Northern Hemisphere band
    lat_mask = (lats >= 30) & (lats <= 70)
    nh_winds = wind_magnitude[lat_mask]
    nh_lats = lats[lat_mask]
    
    jet_lats = np.zeros(len(lons))
    uncertainties = np.zeros(len(lons))
    
    for i in range(len(lons)):
        wind_profile = nh_winds[:, i]
        weights = wind_profile**2
        
        # Calculate weighted mean (jet position)
        jet_lats[i] = np.average(nh_lats, weights=weights)
        
        # Calculate uncertainty based on "peakedness" of wind profile
        max_wind = np.max(wind_profile)
        # Normalize wind profile
        norm_profile = wind_profile / max_wind
        # Calculate spread of high wind speeds (e.g., >80% of max)
        high_winds = norm_profile > 0.8
        uncertainty = np.sum(high_winds) / len(wind_profile)  # More spread = higher uncertainty
        uncertainties[i] = uncertainty
    
    # Smooth both the position and uncertainty
    jet_lats = savgol_filter(jet_lats, window_length=51, polyorder=3)
    uncertainties = savgol_filter(uncertainties, window_length=51, polyorder=3)
    
    # Force matching ends for periodic boundary
    avg_end = (jet_lats[0] + jet_lats[-1])/2
    jet_lats[0] = avg_end
    jet_lats[-1] = avg_end
    
    return jet_lats, uncertainties

# Create _frames directory if it doesn't exist
if not os.path.exists('_frames'):
    os.makedirs('_frames')

# Read both U and V component files
ds_u = xr.open_dataset('/Users/raj/Documents/site/polar-vortex/e9abcdf35d292792804450be5e25658b/u_component_of_wind_0_daily-mean.nc')
ds_v = xr.open_dataset('/Users/raj/Documents/site/polar-vortex/e9abcdf35d292792804450be5e25658b/v_component_of_wind_0_daily-mean.nc')

# Loop through each day
for day in range(len(ds_u.valid_time)):
    # Get U and V data for this day at 250 hPa (index 1)
    u_wind = ds_u.u.isel(valid_time=day, pressure_level=1)
    v_wind = ds_v.v.isel(valid_time=day, pressure_level=1)

    # Find jet stream position and uncertainty
    jet_latitudes, uncertainties = find_jet_stream_with_uncertainty(wind_magnitude, ds_u.latitude, ds_u.longitude)
    
    # Convert uncertainty to line width (inverse relationship)
    max_width = 4
    min_width = 0.5
    line_widths = max_width * (1 - uncertainties)
    line_widths = np.clip(line_widths, min_width, max_width)
    
    # Calculate wind magnitude
    wind_magnitude = np.sqrt(u_wind**2 + v_wind**2)
    
    date = ds_u.valid_time[day].dt.strftime('%Y-%m-%d').values
    
    # Create plot with map projection
    fig, ax = plt.subplots(figsize=(15,10), 
                          subplot_kw={'projection': ccrs.PlateCarree()})
    
    # Add continental outlines
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    
    # Plot wind magnitude
    im = wind_magnitude.plot.contourf(
        levels=20,
        cmap='viridis',
        extend='both',
        transform=ccrs.PlateCarree()
    )
    
    # Find and plot jet stream
    jet_latitudes = find_jet_stream(wind_magnitude, ds_u.latitude, ds_u.longitude)
    ax.plot(ds_u.longitude, jet_latitudes, 'r-', linewidth=2, transform=ccrs.PlateCarree())
    
    plt.title(f'Wind Speed (m/s) at 250 hPa - {date}')
    plt.colorbar(im, label='Wind Speed (m/s)')
    
    # Save plot with padded day number
    plt.savefig(f'_frames/{day+1:02d}.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f'Saved frame {day+1} of {len(ds_u.valid_time)}')
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import os
import numpy as np
from scipy.signal import savgol_filter
import pandas as pd

def find_jet_stream_with_uncertainty(wind_magnitude, lats, lons):
    # Focus on Northern Hemisphere band
    lat_mask = (lats >= 30) & (lats <= 70)
    nh_winds = wind_magnitude[lat_mask]
    nh_lats = lats[lat_mask]
    
    jet_lats = np.zeros(len(lons))
    uncertainties = np.zeros(len(lons))
    
    for i in range(len(lons)):
        wind_profile = nh_winds[:, i]
        weights = wind_profile**2
        
        # Calculate weighted mean (jet position)
        jet_lats[i] = np.average(nh_lats, weights=weights)
        
        # Calculate uncertainty based on "peakedness" of wind profile
        max_wind = np.max(wind_profile)
        # Normalize wind profile
        norm_profile = wind_profile / max_wind
        # Calculate spread of high wind speeds (e.g., >80% of max)
        high_winds = norm_profile > 0.8
        uncertainty = np.sum(high_winds) / len(wind_profile)  # More spread = higher uncertainty
        uncertainties[i] = uncertainty
    
    # Smooth both the position and uncertainty
    jet_lats = savgol_filter(jet_lats, window_length=51, polyorder=3)
    uncertainties = savgol_filter(uncertainties, window_length=51, polyorder=3)
    
    # Force matching ends for periodic boundary
    avg_end = (jet_lats[0] + jet_lats[-1])/2
    jet_lats[0] = avg_end
    jet_lats[-1] = avg_end
    
    return jet_lats, uncertainties

# Create directories if they don't exist
if not os.path.exists('_frames'):
    os.makedirs('_frames')
if not os.path.exists('_data'):
    os.makedirs('_data')

# Read wind and temperature data
ds_u = xr.open_dataset('/Users/raj/Documents/site/polar-vortex/e9abcdf35d292792804450be5e25658b/u_component_of_wind_0_daily-mean.nc')
ds_v = xr.open_dataset('/Users/raj/Documents/site/polar-vortex/e9abcdf35d292792804450be5e25658b/v_component_of_wind_0_daily-mean.nc')
ds_t = xr.open_dataset('/Users/raj/Documents/site/polar-vortex/e9abcdf35d292792804450be5e25658b/856e360e31988e26cc063c15d136d3a0.nc')

# Initialize list to store all jet stream positions
all_jet_positions = []

# Loop through each day
for day in range(31):  # Changed to full month
    # Get U and V data for this day at 250 hPa (index 1)
    u_wind = ds_u.u.isel(valid_time=day, pressure_level=1)
    v_wind = ds_v.v.isel(valid_time=day, pressure_level=1)
    
    # Get temperature data for this day
    temp = ds_t.t.isel(valid_time=day, pressure_level=0)
    
    # Convert temperature from Kelvin to Celsius
    temp = temp - 273.15
    
    # Calculate wind magnitude for jet stream
    wind_magnitude = np.sqrt(u_wind**2 + v_wind**2)
    
    date = pd.to_datetime(ds_u.valid_time[day].values)  # Fixed date conversion
    
    # Create plot with map projection
    fig, ax = plt.subplots(figsize=(15,10), 
                          subplot_kw={'projection': ccrs.PlateCarree()})
    
    # Add continental outlines
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    
    # Plot temperature
    im = temp.plot.contourf(
        levels=np.linspace(-50, 30, 21),
        cmap='RdBu_r',
        extend='both',
        transform=ccrs.PlateCarree()
    )
    
    # Find jet stream position and uncertainty
    jet_latitudes, uncertainties = find_jet_stream_with_uncertainty(wind_magnitude, ds_u.latitude, ds_u.longitude)
    
    # Create DataFrame for this day's positions
    positions_df = pd.DataFrame({
        'date': [date] * len(ds_u.longitude),  # Repeat date for each longitude
        'longitude': ds_u.longitude.values,
        'latitude': jet_latitudes,
        'uncertainty': uncertainties
    })
    all_jet_positions.append(positions_df)
    
    # Convert uncertainty to line width
    max_width = 5
    min_width = 0.1
    line_widths = max_width * (1 - uncertainties)
    line_widths = np.clip(line_widths, min_width, max_width)
    
    # Plot with varying line width
    for i in range(len(ds_u.longitude)-1):
        ax.plot(ds_u.longitude[i:i+2], jet_latitudes[i:i+2], 
                'k-',
                linewidth=line_widths[i],
                transform=ccrs.PlateCarree())
    
    plt.title(f'Surface Temperature (°C) with Jet Stream - {date.strftime("%Y-%m-%d")}')
    plt.colorbar(im, label='Temperature (°C)')
    
    # Save plot
    plt.savefig(f'_frames/{day+1:02d}.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f'Saved frame {day+1} of {len(ds_u.valid_time)}')

# Combine all days into one DataFrame and save
all_positions = pd.concat(all_jet_positions)
all_positions.to_csv('_data/jet_stream_positions.csv', index=False)

print("Saved jet stream positions to _data/jet_stream_positions.csv")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Read the saved positions
all_positions = pd.read_csv('_data/jet_stream_positions.csv')
all_positions['date'] = pd.to_datetime(all_positions['date'])

# Define North American longitude bounds (roughly)
lon_west = 230  # 230°E = 130°W
lon_east = 300  # 300°E = 60°W

def compute_polar_cap_area(group_data, lon_west, lon_east):
    # Filter for North American longitudes
    na_data = group_data[(group_data['longitude'] >= lon_west) & 
                        (group_data['longitude'] <= lon_east)].copy()
    
    # Sort by longitude to ensure proper integration
    na_data = na_data.sort_values('longitude').reset_index(drop=True)
    
    # Convert to radians for calculation
    lats_rad = np.radians(na_data['latitude'].values)
    lons_rad = np.radians(na_data['longitude'].values)
    
    # Earth's radius in kilometers
    R = 6371
    
    # Calculate area using polar cap formula
    # Area = R² * (2π * (1 - sin(θ))) where θ is the latitude
    # For a segment, we multiply by (δλ/2π) where δλ is the longitude span
    
    # Calculate area for each longitude segment
    total_area = 0
    for i in range(len(lons_rad)-1):
        lat = lats_rad[i]
        dlon = lons_rad[i+1] - lons_rad[i]
        area = R**2 * (1 - np.sin(lat)) * dlon
        total_area += area
    
    return total_area

# Calculate area for each day
daily_areas = []
for date, group in all_positions.groupby('date'):
    area = compute_polar_cap_area(group, lon_west, lon_east)
    daily_areas.append({'date': date, 'area': area})

# Convert to DataFrame
daily_areas_df = pd.DataFrame(daily_areas)

# Find the day with maximum area
max_area_day = daily_areas_df.loc[daily_areas_df['area'].idxmax()]

print("\nDaily Areas (km²):")
print(daily_areas_df)
print(f"\nMaximum polar cap area occurred on: {max_area_day['date'].strftime('%Y-%m-%d')}")
print(f"Area: {max_area_day['area']:.0f} km²")

# Optional: Plot the areas over time
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(daily_areas_df['date'], daily_areas_df['area'] / 1e6, 'b-')
plt.title('Polar Cap Area Over Time')
plt.xlabel('Date')
plt.ylabel('Area (million km²)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('_data/polar_cap_area.png')
plt.close()

print("\nArea plot saved as '_data/polar_cap_area.png'")
```
